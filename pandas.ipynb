```python
data = {
    'Name': ['Anish','Kohli','Rahane','Rohit','Virat',
             'Gill','Pant','Hardik','Surya','Jadeja'],

    'Dept': ['IT','IT','CSE','ECE','IT',
             'CSE','ECE','IT','CSE','ECE'],

    'Salary': [40000, 50000, 60000, 45000, 70000,
               52000, 48000, 65000, 55000, 47000],

    'Experience_Years': [1, 3, 7, 4, 10,
                         5, 4, 8, 6, 5],

    'Location': ['Bangalore','Mumbai','Pune','Chennai','Delhi',
                 'Pune','Chennai','Mumbai','Bangalore','Delhi'],

    'Performance_Rating': [3.5, 4.0, 4.5, 3.8, 4.9,
                           4.2, 3.9, 4.6, 4.1, 3.7]
}

data = pd.DataFrame(data)
data
```
