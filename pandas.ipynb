Joins


```python
df_employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104, 105, 106],
    'Name': ['Anish', 'Kohli', 'Rahane', 'Rohit', 'Virat', 'Gill'],
    'Dept': ['IT', 'IT', 'CSE', 'ECE', 'IT', 'CSE'],
    'Location': ['Bangalore', 'Mumbai', 'Pune', 'Chennai', 'Delhi', 'Pune'],
    'Join_Year': [2022, 2021, 2019, 2020, 2018, 2021]
})

df_employees
```

```python
df_salary = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 105, 107],
    'Salary': [40000, 50000, 60000, 70000, 45000],
    'Bonus': [5000, 6000, 8000, 10000, 4000],
    'Currency': ['INR', 'INR', 'INR', 'INR', 'INR'],
    'Pay_Grade': ['B1', 'B2', 'B3', 'B4', 'B2']
})


df_salary
```

```python
df_performance = pd.DataFrame({
    'Emp_ID': [101, 101, 102, 103, 103, 105],
    'Review_Year': [2023, 2024, 2024, 2023, 2024, 2024],
    'Rating': [3.5, 4.0, 4.2, 4.6, 4.8, 4.9],
    'Promotion_Eligible': ['No', 'Yes', 'Yes', 'No', 'Yes', 'Yes'],
    'Reviewer': ['Mgr1', 'Mgr2', 'Mgr1', 'Mgr3', 'Mgr2', 'Mgr1']
})

df_performance
```

```python
df_new_hires = pd.DataFrame({
    'Emp_ID': [108, 109],
    'Name': ['Pant', 'Hardik'],
    'Dept': ['ECE', 'IT'],
    'Location': ['Chennai', 'Mumbai'],
    'Join_Year': [2024, 2024]
})

df_new_hires
```

```python
#finding emp name, id, dept and their salries

df_emp_sal = pd.merge(
    df_employees[['Emp_ID','Name','Dept']],
    df_salary[['Emp_ID','Salary']],
    on = 'Emp_ID',
    how = 'inner'
)

df_emp_sal
```

```python
#LEFT JOIN

df_emp_sal = pd.merge(
    df_employees[['Emp_ID','Name','Dept']],
    df_salary[['Emp_ID','Salary']],
    on = 'Emp_ID',
    how = 'left'
)

df_emp_sal
```

```python
df_emp_sal.isnull().sum()
```

```python
df_emp_sal.fillna(df_salary['Salary'].mean())
```

```python
#RIGHT JOIN

df_emp_sal = pd.merge(
    df_employees[['Emp_ID','Name','Dept']],
    df_salary[['Emp_ID','Salary']],
    on = 'Emp_ID',
    how = 'right'
)

df_emp_sal
```

CONCAT

```python
df_full_emp = pd.concat(
    [df_employees,df_new_hires],
    join='outer',
    ignore_index=True
)

df_full_emp
```
