```python
emp_list=[('Ankit',10000),('Rahul',12000),('Sumit',14000),('Dheeraj',21000),('Pavan',11000),('Mohit',13000)]

avg_sal = 0
for name,salary in emp_list:
    avg_sal = (avg_sal + salary)

average_salary = (avg_sal)/len(emp_list)
print(average_salary)

new_list = []
for name,salary in emp_list:
    if salary > average_salary:
        new_list.append([name,salary])
print(new_list)
```
